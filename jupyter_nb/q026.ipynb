```python
# 기능체크-순환부분
# 루프문으로 여러사이즈의 순환마디를 확인.

string2='14285714285714285'
for i in range(len(string2)):
    part1=string2[:i]
    part2=string2[i:i*2]
    print("lopp:",i,part1,part2,end=',')
    if part1 == part2:
        print(part1,"is reciprocal cycle")
    print()
    
```

```python
# 메인루프-1 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i+1,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if part1 == part2:
                cycle_string=part1
    return cycle_string

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-2 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류 -> close-> new Issue 3.
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29
# Issue 3: 한자리 순환처리

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string)): #Issue-1
                cycle_string=part1
    return cycle_string

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-3 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류 -> close -> new Issue 3.
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29 -> close
# Issue 3: 한자리 순환처리

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string)  #Issue -1
                and len(part1)*string.count(part1)>len(string)/2): #Issue -2
                cycle_string=part1
    return cycle_string

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-4 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류 -> close -> new Issue 3.
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29 -> close
# Issue 3: 한자리 순환처리 -> close

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string) #Issue -1
                and len(part1)*string.count(part1)>len(string)/2): #Issue-2
                cycle_string=part1
    #Issue-3 
    cycle_string_min=cycle_string
    for i in range(len(cycle_string)):
        part1=cycle_string[:i]
        if (cycle_string.count(part1)*len(part1)==len(cycle_string)
            and len(part1)<len(cycle_string_min)):
            cycle_string_min=part1
    
    return cycle_string, cycle_string_min

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-6 (decimal 1000, number 100까지)
# decimal 10,100,1000 별로 순환마디 체크/비교.

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string) #Issue -1
                and len(part1)*string.count(part1)>len(string)/2): #Issue-2
                cycle_string=part1
    #Issue-3 
    cycle_string_min=cycle_string
    for i in range(len(cycle_string)):
        part1=cycle_string[:i]
        if (cycle_string.count(part1)*len(part1)==len(cycle_string)
            and len(part1)<len(cycle_string_min)):
            cycle_string_min=part1
    return len(cycle_string_min)

getcontext().prec = 10
result=[]
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    result.append([d,getReciprocal(string)])
    if result[d-1][1] != 0:
        print(result[d-1])

getcontext().prec = 100
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    result[d-1].append(getReciprocal(string))
    if result[d-1][1] != result[d-1][2]:
        print(result[d-1])
        
getcontext().prec = 1000
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    result[d-1].append(getReciprocal(string))
    if result[d-1][2] != result[d-1][3]:
        print(result[d-1])
```
