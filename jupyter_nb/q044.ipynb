# Q044 Pentagon numbers


Pentagonal numbers are generated by the formula, P$_n$=n(3n−1)/2. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P$_4$ + P$_7$ = 22 + 70 = 92 = P$_8$. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, P$_j$ and P$_k$, for which their sum and difference are pentagonal and D = |P$_k$ − P$_j$| is minimised; what is the value of D?

```python
#forum-8
def p44():
    pent, bool_pent, end = [0], [0], 1

    while True:
        start, end = end, end*2

        for n in xrange(start, end):
            pent += [n * (3*n - 1) / 2]
            bool_pent += 3 * (n-1) * [0] + [1]
        
        upper_bound = len(bool_pent) - 1
        for j in xrange(1, end):
            for k in xrange(j+1, end):
                diff, sum = pent[k] - pent[j], pent[k] + pent[j]
                if sum <= upper_bound and bool_pent[diff] and bool_pent[sum]:
                    return diff
```

```python
#forum-10
import time

def is_pentagonal(a):
    b = ((24 * a + 1) ** 0.5 + 1) / 6

    if int(b) == b:
        return True

    else:
        return False

def result():
    pentagonal_list = [int(0.5 * x * (3 * x - 1)) for x in range(1, 2200)]

    for i in range(1, 1200):
        for j in range(i, len(pentagonal_list)):
            first_number  = pentagonal_list[j]
            second_number = pentagonal_list[j - i]

            sum_temp = first_number + second_number
            dif_temp = first_number - second_number

            if is_pentagonal(sum_temp) and is_pentagonal(dif_temp):
                return dif_temp

start_time = time.time()

print(result())
print("{} ms".format(round(1000 * (time.time() - start_time))))
```
