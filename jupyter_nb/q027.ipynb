# Q027 Quadratic primes

Euler discovered the remarkable quadratic formula:

$$n^2+n+41$$

It turns out that the formula will produce 40 primes for the consecutive integer values 0≤n≤39. However, when $n=40,40^2+40+41=40(40+1)+41$ is divisible by 41, and certainly when $n=41,41^2+41+41$ is clearly divisible by 41.

The incredible formula $n^2−79n+1601$ was discovered, which produces 80 primes for the consecutive values 0≤n≤79. The product of the coefficients, −79 and 1601, is −126479.

Considering quadratics of the form:

$$n^2+an+b, where |a|<1000 and |b|≤1000$$

where |n| is the modulus/absolute value of n
$$e.g. |11|=11 and |−4|=4$$

Find the product of the coefficients, a and b, for the quadratic expression that produces the maximum number of primes for consecutive values of n, starting with n=0.

```python
#forum-1
from itertools import count
from math import sqrt

from bench import measure  # measures run time
from pe_10 import sieve  # a sieve of erath…, first used in problem 10


def prime_test(n):
    if n < 2:
        return False
    for i in range(2, 1 + int(sqrt(n))):
        if n % i == 0:
            return False
    return True


def solve(bound):
    primes = tuple(sieve(bound + 1))
    b_candidates = tuple(reversed(primes))
    not_primes = set()
    primes = set(primes)
    best_ab, best_score = (0, 0), 0
    for b in b_candidates:
        for a in range(0, -bound, -1):
            for n in count(0):
                v = n * n + n * a + b
                is_prime = v in primes
                if not is_prime and v > bound and v not in not_primes:
                    is_prime = prime_test(v)
                    if is_prime:
                        primes.add(v)
                    else:
                        not_primes.add(v)
                if not is_prime:
                    if n > best_score:
                        best_score = n
                        best_ab = a, b
                    break
    a, b = best_ab
    return a * b


if __name__ == '__main__':
    print(measure(lambda: solve(1000)))
```
