# Permuted multiples


It can be seen that the number, 125874, and its double, 251748, contain exactly the same digits, but in a different order.

Find the smallest positive integer, x, such that 2x, 3x, 4x, 5x, and 6x, contain the same digits.


```python
# check example

def is_same_digit(string_cmp):
    for i in string:
        if i not in string_new:
            return False
    return True

number = 125874
string = str(number)
print(string)

for i in range(2,7):
    string_new = str(number*i)
    print(string_new)
    if not len(string) == len(string_new):
        print("false")
        break
    elif is_same_digit(string_new):
            print("is same digit")
    else:
        print("not same")
        break
```

```python
# ~100

def is_same_digit(string_cmp):
    for i in string:
        if not string.count(i) == string_cmp.count(i):
            return False
    return True

for number in range(1,100):
    string = str(number)
    count = 0
    for i in range(2,7):
        string_new = str(number*i)
        if not len(string) == len(string_new):
            print("loop:",string,string_new,"length different")
            break
        elif is_same_digit(string_new):
                count+=1
                print("loop:",string,string_new,"is same digit","count:",count)
        else:
            print("loop:",string,string_new,"not same digit, ","count:",count)
            break

    if count == 5:
        print(number,"is answer")


```

```python
# sol : ~ Infinite

def is_same_digit(string_cmp):
    for i in string:
        if not string.count(i) == string_cmp.count(i):
            return False
    return True

number=1
while True:
    string = str(number)
    count = 0
    for i in range(2,7):
        string_new = str(number*i)
        if not len(string) == len(string_new):
            break
        elif is_same_digit(string_new):
                count+=1
        else:
            break
    if count >1:
        print("loop:",string,string_new,"is same digit","multiply:",i,"count:",count)
    if count == 5:
        print(number,"is answer")
        break
    number+=1


```

```python
# forum -1

def f(n):
    s = str(n)
    l = []
    for i in range(len(s)):
        l.append(s[i:i+1])
    m = sorted(l)
    t = ""
    for d in m:
        t += d
    return t

x = 1
while True:
    if f(x) == f(2*x) and f(x) == f(3*x) and f(x) == f(4*x) and f(x) == f(5*x) and f(x) == f(6*x):
        print(x)
        break
    x += 1

```

```python
# forum -2
def SameDigit(exlist):
    for i in  exlist:
        if set(i)!=set(exlist[0]):
            return False
    return True

Number=1
while 1:
    Nlist=[str(i*Number) for i in range(1,7)]
    if SameDigit(Nlist):
        print(Nlist)
        break
    else:
        Number+=1
```
