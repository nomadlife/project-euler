# Q049 Prime Permutations


The arithmetic sequence, 1487, 4817, 8147, in which each of the terms increases by 3330, is unusual in two ways: (i) each of the three terms are prime, and, (ii) each of the 4-digit numbers are permutations of one another.

There are no arithmetic sequences made up of three 1-, 2-, or 3-digit primes, exhibiting this property, but there is one other 4-digit increasing sequence.

What 12-digit number do you form by concatenating the three terms in this sequence?

```python
#forum-3
import math
import bisect as bi

def isPermutation(n, m):
    n1 = sorted(str(n))
    m1 = sorted(str(m))
    return m1==n1

def isPrime(n, lstPrimes):
    m = math.sqrt(n)
    for i in lstPrimes:
        if i > m:
            return True
        if n % i == 0:
            return False
    return True

def getPrimes(lowerBound, upperBound):
    lst = [2,3]
    i = 5
    inc = 2
    while i <= upperBound:
        if isPrime(i, lst):
            lst.append(i)
        i += inc
        inc = 6 - inc
    lst = filter(lambda x: x >= lowerBound, lst)
    return lst


def index(a, x):
    'Locate the leftmost value exactly equal to x'
    i = bi.bisect_left(a, x)
    if i != len(a) and a[i] == x:
        return i
    return -1

def p49(primes, delta):
    for p in primes:
        i1 = index(primes, p+delta)
        i2 = index(primes, p+2*delta)
        if p!=1487 and i1!=-1 and i2!=-1:
            if isPermutation(p, primes[i1]) and isPermutation(p, primes[i2]):
                return p

primesOfInterest = getPrimes(1000, 9999)
p = p49(primesOfInterest, 3330)
```

```python
#forum-5
def primes_under(n):
    prime_list = [i for i in range(2, n)]
    for i in range(2, n):
        for j in range(i ** 2, n, i):
            if j in prime_list:
                prime_list.remove(j)
    return prime_list


def prime_check(n):
    n = int(n)
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


def arithmetic_three_permutations(a, b):
    f = [int(i) for i in str(a)]
    g = [int(i) for i in str(a + b)]
    h = [int(i) for i in str(a + 2 * b)]
    if sorted(f) == sorted(g) == sorted(h):
        if prime_check(a) and prime_check(a + b) and prime_check(a + 2 * b) is True:
            return True
    return False


x = primes_under(10000)
for i in x:
    if i > 999:
        for j in range(2, 4500, 2):
            if arithmetic_three_permutations(i, j) is True:
                print("".join([str(i), str(i + j), str(i + 2 * j)]))
                
```

```python
#forum-9
def list_is_prime(n, primes_too=False):
    if n <= 0:
        return [] if n else [False]

    primes = [2, 3]
    for y in xrange(5, n + 1, 6):
        for x in [y, y + 2] if y + 2 <= n else [y]:
            square = int(sqrt(x))
            for p in primes:
                if p > square:
                    primes += [x]
                    break
                if x % p == 0:
                    break

    bool_primes = [False]*(n+1)

    for p in primes:
        bool_primes[p] = True

    if primes_too:
        return bool_primes, primes
    return bool_primes


def p49():
    primes = list_is_prime(10000)

    for x in xrange(1000, 10000):
        sorted_x = sorted(str(x))
        if primes[x] and x != 1487:
            for jump in xrange(1, (10000 - x) / 2):
                x1, x2 = x+jump, x+2*jump
                if primes[x1] and primes[x2]:
                    if sorted(str(x2)) == sorted_x and sorted(str(x1)) == sorted_x:
                        return '%d%d%d\n' % (x, x1, x2)
```
