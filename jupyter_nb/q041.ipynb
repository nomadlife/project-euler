```python
#forum-11
import time

def is_prime(a):
    if a == 1:
        return False

    for b in range(2, round(a ** 0.5) + 1):
        if a % b == 0:
            return False

    return True

def permutation(number):
    final_list = []

    if len(number) == 1:
        final_list.append(number)

    else:
        for c in range(len(number)):
            permutation_list = permutation(number[0:c] + number[c + 1:len(number)])

            for d in permutation_list:
                final_list.append(number[c] + d)

    return final_list

def result():
    number_list = "123456789"

    for i in range(len(number_list)):
        temporary   = number_list[:len(number_list) - i]
        all_numbers = permutation(temporary)
        all_numbers.sort(reverse = True)

        for j in range(len(all_numbers)):
            check = int(all_numbers[j])

            if check % 2 == 0 or not is_prime(check):
                continue

            else:
                return check

start_time = time.time()

print(result())
print("{} ms".format(round(1000 * (time.time() - start_time))))
```
