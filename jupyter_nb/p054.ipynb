## Poker hands




```python
with open('p054_poker.txt') as data:
    lines = data.readlines()

cards=[]
for line in lines:
    Arr = line.split()
    p1n,p1s,p2n,p2s ='','','',''
    for i in range(5):
        p1n += Arr[i][0] # player1 card numbers
        p1s += Arr[i][1] # player1 card suits
        p2n += Arr[i+5][0] # player2 card numbers
        p2s += Arr[i+5][1] # player2 card suits
    cards.append([p1n,p1s,p2n,p2s])
#print(cards)
```

```python
def getScore(numbers, suits):
    newStr = ''; score =0; score2nd=''; score3rd=''
    for i in 'AKQJT98765432':
        count = numbers.count(i)
        if count > 0: 
            if count == 1 : score3rd += i
            elif count == 2 : score += 3; score2nd += i # pair
            elif count == 3 : score += 9; score2nd = i + score2nd # Three
            elif count == 4 : score += 15; score2nd = i # 4card
            newStr += i*count
    if newStr in 'AKQJT98765432':
        score += 10; score2nd += newStr[0] # straight
    if all(suits[i] == suits[i+1] for i in range(4)): 
        score += 11; score2nd += suits[0] # flush
    return score, score2nd, score3rd

print(getScore('82346','HSDDH'))
```

```python
count=0
for i in range(len(cards)):
    numStr = '23456789TJQKA'
    p1 = getScore(cards[i][0],cards[i][1])
    p2 = getScore(cards[i][2],cards[i][3])
    winner = ''
    if p1[0] > p2[0]: 
        winner = 'player1s1'; count+=1
    elif p1[0] < p2[0]: 
        winner = 'player2s1'
    elif winner=='' and len(p1[1]) >= 1 and numStr.index(p1[1]) > numStr.index(p2[1]):
        winner = 'player1s2'; count+=1
    elif len(p1[2]) > 1 :
        for j in range(len(p1[1])):
            if numStr.index(p1[2][j]) > numStr.index(p2[2][j]):
                winner = 'player1s3'; count+=1
                break
    else : winner = 'player2'
    print('{} {:2} {:2} {:2} {:2} {:5} {:5} {}'
          .format(lines[i].rstrip(), p1[0], p2[0], p1[1], p2[1],p1[2],p2[2], winner))
print(count)
```

```python
def calc(numbers, suits):
    result = {
    'newStr':'',
    'rank':'',
    'score':0,
    'score2nd':'',
    'score3rd':''
    }
    for i in 'AKQJT98765432':
        count = numbers.count(i)
        if count > 0: 
            if count == 2 : result['score'] += 3; result['score2nd'] += i # pair
            elif count == 3 : result['score'] += 9; result['score2nd'] = i + result['score2nd'] # Three
            elif count == 4 : result['score'] += 15; result['score2nd'] = i # 4card
            result['newStr'] += i*count
    if result['newStr'] in 'AKQJT98765432':
        result['score'] += 10; result['score2nd'] += result['newStr'][0] # straight
    if all(suits[i] == suits[i+1] for i in range(4)): 
        result['score'] += 11; result['score2nd'] += suits[0] # flush
    return result

print(calc('33534','HSDDH'))
```

```python
def findWinner(p1, p2):
    rankDict = {
        'Straight Flush':21,
        'Four of a Kind' : 15,
        'Full House' : 12,
        'Flush': 11,
        'Straight':10,
        'Three of a Kind' : 9,
        'Two Pairs' : 6,
        'One Pair' : 3
    }
    if p1['score'] > p2['score']: return 'player1'
    elif p1['score'] < p2['score']: return 'player2'
    else : return 'tie'
```

```python

count=0
for i in range(len(cards)):
    p1 = calc(cards[i][0],cards[i][1])
    p2 = calc(cards[i][2],cards[i][3])
    winner = findWinner(p1,p2)
    print('{}  {:2}  {:2}  {}'.format(lines[i].rstrip(), p1['score2nd'], p2['score2nd'], winner))
print(count)
```
