```python
#forum-2
from math import sqrt


def mcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m


def triangles(max_perimeter=1000):
    nt = [0 for i in range(max_perimeter + 1)]  # nt[i] number of triangles with perimeter [i]

    for m in range(2, int(sqrt(max_perimeter / 2))):  # is an lazy upper bound. with 2m^2 > max_perimeter
        for n in range(1, m):
            if ((m + n) % 2 == 1) and mcd(m, n) == 1:
                base = 2 * m * (m + n)
                index = base
                while index <= max_perimeter:
                    nt[index] += 1
                    index += base
    return nt

a = triangles()

print(a.index(max(a)))
```

```python
#forum-3
import math
from collections import Counter
from time import time


def checkPS(x):
    if math.sqrt(x) == math.floor(math.sqrt(x)): return True
    else: return False

def most_common(lst):
    data = Counter(lst)
    return max(lst, key=data.get)

timeThen = time()
p_s=[]

for c in range(5,500):
    for b in range(c//2,c):
        if checkPS(c*c - b*b):
            a = math.floor(math.sqrt(c*c - b*b))
            p = [a,b,c,a+b+c]
            if p[1]>p[0] and p[3]<=1000: p_s.append(p[3])

print(most_common(p_s))
print(time()-timeThen,'seconds')
```

```python
#forum-4
def p39():
    max_p, max_sol, n = 0, 0, 1000

    for p in xrange(n + 1):
        sols = 0

        for c in xrange(5, p/2):
            delta = 4*c**2 + 8*p*c - 4*p**2
            if delta >= 0 and sqrt(delta) == int(sqrt(delta)):
                sols += 1

        if sols > max_sol:
            max_sol = sols
            max_p = p

    return max_p
```

```python
#forum-5
import time

def result(n):
    sum_list = []

    for x in range(1, n):
        for y in range(x, n):
            a = x ** 2 + y ** 2
            z = int(a ** 0.5)

            if a / z == z:
                total_sum = x + y + z

                if total_sum <= n:
                    sum_list.append(total_sum)

    return max(sum_list, key = sum_list.count)

start_time = time.time()

print(result(1000))
print("{} ms".format(round(1000 * (time.time() - start_time))))
```

```python
#forum-6
def integer_right_triangles(target):
    result = {}
    c = 5
    a = 1
    while a <= target // 2 - 1:
        b = 1
        while b <= target // 2 - a:
            c = (a ** 2 + b ** 2) ** 0.5
            if c == int(c):
                add = (max(a, b), min(a, b), int(c))
                p = int(a + b + c)
                if p <= target:
                    if p not in result.keys():
                        result[p] = [add]
                    else:
                        if add not in result[p]:
                            result[p].append(add)
            b += 1
        a += 1
    max_value = 0
    max_key = 0
    for (key, value) in result.items():
        if len(value) > max_value:
            max_value = len(value)
            max_key = key
    return max_key, max_value

print(integer_right_triangles(1000

))
```

```python
#forum-7
def sides(p):
    for c in range(1,p+1):
        for b in range(1,c+1):
            a = p-c-b
            if (c*c==b*b+a*a and a+b+c == p):
                 yield [a,b,c]
max = 0
maxP = 0
for p in range (1,1001):
    l = []
    for i in sides(p):
        l.append(i)
    #print("{0} {1}".format(p, l))

    if len(l) > max:
        max = len(l)
        maxP = p
print(maxP)
```
