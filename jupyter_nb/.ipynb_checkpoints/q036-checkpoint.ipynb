```python
#forum-7
def isPalindromic(n):
  if str(n) == str(n)[::-1]:
    return True
  return False
  
def toBinary(n):
  return "{0:b}".format(n)
  
def main():
  sum = 0
  for i in range(1000000):
    if isPalindromic(i) and isPalindromic(toBinary(i)):
      sum+=i
  print(sum)

main()
```

```python
#forum -8

def isPalinNum(num):                              # is it palindrome?
	numStr = str(num)
	for i in range(1, int(len(numStr)/2)+1):
		if numStr[i-1] != numStr[-i]:
			return False
	return True


#########################################

numSet = set()
for num in range(1, 1000001):
	numStr = str(num)

	if numStr[-1] == '0':
		continue

	if isPalinNum(numStr):
		binNum = str(bin(num)[2:])
		if isPalinNum(binNum):
			numSet.add(num)
			print('good')

print(sum(numSet), sorted(numSet))
```

```python
#forum-10
##Palindrome numbers for 2 Digits, 3 Digits, etc.
D2 = [11, 22, 33, 44, 55, 66, 77, 88, 99]
D3 = []
D4 = []
D5 = []
D6 = []

def PrintD(D):
    for i in range (0, len(D)):
        print(D[i])

def fill3D():
    for i in range(0, len(D2)):
        for j in range(0, 10):
            D3.append(int(str(D2[i])[0] + str(j) + str(D2[i])[1]))

def fill4D():
    for i in range(0, len(D2)):
        for j in range(0, 10):
            D4.append(int(str(D2[i])[0] + str(j) + str(j) + str(D2[i])[1]))

def fill5D():
    for i in range(0, len(D2)):
        for j in range(0, 10):
            D5.append(int(str(D2[i])[0] + "0" + str(j) + "0" + str(D2[i])[1]))
        for j in range(0, len(D3)):
            D5.append(int(str(D2[i])[0] + str(D3[j]) + str(D2[i])[1]))

def fill6D():
    for i in range(0, len(D2)):
        for j in range(0, 10):
            D6.append(int(str(D2[i])[0] + "0" + str(j) + str(j) + "0" + str(D2[i])[1]))
        for j in range(0, len(D3)):
            D6.append(int(str(D2[i])[0] + str(D4[j]) + str(D2[i])[1]))

fill3D()
fill4D()
fill5D()
fill6D()
#print("\n D6: ")
#PrintD(D6)

##returns true if binary version of num is a palindrome
def CheckBinaryPali(num):
    a = '{0:b}'.format(num)
    b = ''.join(reversed(a))
    return a == b


def FindDoublePalindromes():
    sum = 0

    for i in range(0, 10):
        if CheckBinaryPali(i):
            sum += i

    for i in range(0, len(D2)):
        if CheckBinaryPali(D2[i]):
            sum += D2[i]

    for i in range(0, len(D3)):
        if CheckBinaryPali(D3[i]):
            sum += D3[i]

    for i in range(0, len(D4)):
        if CheckBinaryPali(D4[i]):
            sum += D4[i]

    for i in range(0, len(D5)):
        if CheckBinaryPali(D5[i]):
            sum += D5[i]

    for i in range(0, len(D6)):
        if CheckBinaryPali(D6[i]):
            sum += D6[i]

    return sum

print("\n\n Answer: " + str(FindDoublePalindromes()))
```
