```python
# dictionary -5 
eqn = {
    3 : 'n*(n+1)/2',
    4 : 'n**2',
    5 : 'n*(3*n-1)/2',
    6 : 'n*(2*n-1)',
    7 : 'n*(5*n-3)/2',
    8 : 'n*(3*n-2)'
}          
numbers = {} 
for i in range(3,9):
    for n in range(1,200):
        val = eval(eqn[i])
        if val >= 10000: break
        elif val > 1000: 
            if (val//100) in numbers:
                numbers[val//100] = numbers[val//100] + [(val%100,i,n)]
            else:
                numbers[val//100] = [(val%100,i,n)]

def func(key, string = '', history = ''):
    if len(string) == 6 :
        print(string, history)
        #return
    for i in numbers[key]:
        if(str(i[1]) not in string):
            string+=str(i[1])
            history+=str(i)
            print(key,i,i[0], string, history)
            func(i[0], string, history )
        
for k in numbers.keys():
    print('loop:',k)
    func(k)
```
