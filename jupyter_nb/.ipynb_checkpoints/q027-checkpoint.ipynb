```python
#(수학적 특성 확인) 결과값이 소수인지 확인.
for n in range(40):
    value=pow(n,2)+n+41
    print(value,end=' ')
    prime=True
    for m in range(2,n):
        if value%m==0:
            prime=False
            break
    if prime:
        print(prime)
print(0)

#범위내 모든 소수인지도 확인.
for i in range(41,1602):
    prime=True
    for j in range(2,i):
        if i%j==0:
            prime=False
            break
    if prime:
        print(i,end=' ')
        
```

```python
#(수학적 특성 확인)함수로 생성된 소수와 범위내 소수 갯수차이 확인.
# List 에 저장해서 확인
primeList=[]
for n in range(40):
    value=pow(n,2)+n+41
    print(value,end=' ')
    prime=True
    for m in range(2,n):
        if value%m==0:
            prime=False
            break
    if prime:
        print("prime")
        primeList.append(value)
print(primeList)

primeList2=[]
for i in range(41,1602):
    prime=True
    for j in range(2,i):
        if i%j==0:
            prime=False
            break
    if prime:
        print(i,end=' ')
        primeList2.append(i)
print()
print(primeList2)

for v in primeList2:
    if v in primeList:
        print(v,v)
    else :
        print(v)
```

```python
#단순계산

maxN=0;find_a=9999;find_b=9999
for a in range(-1000,1001):
    print("loop:",a)
    for b in range(-1000,1001):
        n=0;loop=True
        while loop:
            value=pow(n,2)+n*a+b
            if value<=0:
                loop=False
            for i in range(2,value):
                if value%i==0:
                    loop=False
                    break
            if loop==False and n>maxN:
                    maxN=n;find_a=a;find_b=b
                    print("b=",b,n,value,"new max value n")
            n+=1
print("maxN:",maxN,"find_a=",find_a,"find_b=",find_b,"solution=",find_a*find_b)
```
