# Q012 Highly divisible triangular number


The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?


```python
def count_factors(num):
    loop=num**0.5
    # for square number
    if loop==int(loop):
        count=1
    else:
        count=0
    i=1
    while i<loop:
        if num%i==0:
            print(i,end=',')
            count+=2
        i+=1
    print()
    return count

i=1;tri_num=0
while i<100:
    tri_num = tri_num + i
    count = count_factors(tri_num)
    print("loop:",i,"tri_num",tri_num,"count",count)
    i+=1
    print()
```

```python
#forum-1
from time import time
from itertools import count
from math import sqrt


def all_factors(n):
    yield 1
    yield n
    for i in range(2, 1 + int(sqrt(n))):
        d, m = divmod(n, i)
        if m == 0:
            yield i
            yield d


def test_triangle_number(n):
    if n % 2 == 0:
        a = n + 1
        b = n // 2
    else:
        a = n
        b = (n + 1) // 2

    factors = set()
    b_factors = tuple(all_factors(b))
    for x in all_factors(a):
        for y in b_factors:
            factors.add(x * y)
    if len(factors) > 500:
        return a * b
    else:
        return None

start_time = time()
solution = next(filter(None, map(test_triangle_number, count(1))))
print(f'{int(10 ** 3 * (time() - start_time))} ms')
print(solution)
```

```python
#forum-2
import math


def number_of_divisors(number):
    nod = 0
    sqr = int(math.sqrt(number))
    for j in range(1, sqr+1):
        if number % j == 0:
            nod += 2
    if sqr * sqr == number:
        nod -= 1
    return nod


def first_triangle_with_divisors_overs(over):
    number = 0
    i = 1
    while number_of_divisors(number) < over:
        number += i
        i += 1
    return number


over = 500
print(first_triangle_with_divisors_overs(over))
```

```python
#forum-3
from math import sqrt
from numba import jit

@jit
def sum_of_(n):
    return int(n*(n+1)/2)
@jit
def append_LIST(n):
    LIST = []
    for i in range(1,n+1):
        LIST.append(sum_of_(i))
    return LIST
@jit
def check_number_of_divisors(n):
    divs = [1]
    for i in range(2,int(sqrt(n))+1):
        if n%i == 0:
            divs.extend([i,n/i])
    divs.extend([n])
    return len(divs)

if __name__=="__main__":
    LIST = append_LIST(13000)
    for i in LIST:
        if check_number_of_divisors(i)>=500:
            print(i)
            break
```

```python
#forum-4
def generate_primes(max):
    global primes
    primes = []
    primes_list = []
    for i in range(0, max+1):
        primes_list.append(True)
    for i in range(2, max+1):
        if primes_list[i]:
            for j in range(i*2, max+1, i):
                primes_list[j] = False
    for index, i in enumerate(primes_list):
        if index >= 2 and i:
            primes.append(index)


def get_divisors_count(number):
    divisors = get_prime_divisors(number)
    divisors_count = 1
    for prime, amount in divisors.iteritems():
        divisors_count *= amount + 1
    return divisors_count

def get_prime_divisors(number):
    prime_divisors = {}
    global primes
    for prime in primes:
        if number == 1:
            return prime_divisors
        while True:
            if number % prime == 0:
                prime_divisors[prime] = 1 + (prime_divisors[prime] if prime_divisors.has_key(prime) else 0)
                number = number / prime
            else:
                break
    return prime_divisors

def find_answer(limit):
    position = 1
    number = 1
    while True:
        divisors = get_divisors_count(number)
        if divisors >= limit:
            return number
        position += 1
        number += position

generate_primes(100)
print find_answer(499)
```

```python
#forum-6
def divisors(n):
    """
    Find the proper divisors of n.

    Arguments:
        n: the number of which the proper divisors have to be found.

    Return:
        A list containing the divisors of n (including 1 and n itself) sorted in
        ascending order.
    """
    div = 1
    divs = []
    limit = n**.5
    while div < limit:
        if n % div == 0:
            divs.append(div)
            divs.append(n//div)
        div += 1
    return sorted(divs)

def Euler_12(n=500):
    """
    Find the first triangular number with more than n divisors.

    Euler Problem 12: The sequence of triangle numbers is generated by adding
    the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5
    + 6 + 7 = 28. The first ten terms would be:

    1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

    Let us list the factors of the first seven triangle numbers:

    [snip]

    We can see that 28 is the first triangle number to have over five divisors.

    What is the value of the first triangle number to have over five hundred
    divisors?

    Return:
        The first triangular number with more than n divisors.

    """
    triangular = 1
    natural = 2
    while True:
        divs = len(divisors(triangular))
        if divs > n:
            return triangular
        triangular += natural
        natural += 1
```
