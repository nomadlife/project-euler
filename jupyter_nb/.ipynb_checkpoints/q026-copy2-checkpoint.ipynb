# Q026
Find the value of d < 1000 for which 1/d contains the longest recurring cycle in its decimal fraction part.
1000 까지의 수중, 역수의 순환소수부분이 가장 긴 수.


```python
# 기능체크 : 소수점 생성(float 타입), string 처리
# 순환소수가 긴것도 많아서, float로는 처리가 어려움.
value=1/7
print(value,type(value))
string=str(value)
print(string)
string=string[string.find(".")+1:]
print(string)
```

```python
# 기능체크 : 소수점 생성,decimal 타입(긴 소수점 생성 가능)
from decimal import *
getcontext().prec = 100   #소수점이하 100자리까지 설정.
#context = Context(prec=200)
#setcontext(context)
print(Decimal(1)/Decimal(7)) # -> 이표현식 사용.
print(Decimal(1/7))   # 사용불가.
```

```python
# 기능체크 - 순환부분 확인
# 왼쪽부터 순차적으로 잘라서, 나머지 오른쪽부분에서 존재유무 검사. 있으면 순환마디.

string2='14285714285714285'
print(len(string2))
print(string2[0:5] in string2[10:])
```

```python
# 기능체크-순환부분
# 루프문으로 여러사이즈의 순환마디를 확인.

string2='14285714285714285'
for i in range(len(string2)):
    part1=string2[:i]
    part2=string2[i:i*2]
    print("lopp:",i,part1,part2,end=',')
    if part1 == part2:
        print(part1,"is reciprocal cycle")
    print()
    
```

```python
# 기능체크-순환부분
# 첫숫자부터 순환하지 않는 수들이 있어서, 루프문 중첩.
# 결과를 보면, 순환마디가 여러가지로 나오는데, 이숫자에서는 가장 짧은 3이 선택되도록 해야함

string='083333333333'
for i in range(len(string)):
    for j in range(i,len(string)):
        part1=string[i:j]
        part2=string[j:j*2-i]
        print("loop:",i,part1,part2,end=',')
        if part1 == part2:
            print(part1,"is reciprocal cycle")
        print()
```

```python
# 메인루프-1 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i+1,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if part1 == part2:
                cycle_string=part1
    return cycle_string

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-2 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류 -> close-> new Issue 3.
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29
# Issue 3: 한자리 순환처리

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string)): #Issue-1
                cycle_string=part1
    return cycle_string

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-3 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류 -> close -> new Issue 3.
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29 -> close
# Issue 3: 한자리 순환처리

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string)  #Issue -1
                and len(part1)*string.count(part1)>len(string)/2): #Issue -2
                cycle_string=part1
    return cycle_string

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-4 (소수점 100자리, 50까지)
# Issue 1: loop 7,13 등 오류 -> close -> new Issue 3.
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29 -> close
# Issue 3: 한자리 순환처리 -> close

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string) #Issue -1
                and len(part1)*string.count(part1)>len(string)/2): #Issue-2
                cycle_string=part1
    #Issue-3 
    cycle_string_min=cycle_string
    for i in range(len(cycle_string)):
        part1=cycle_string[:i]
        if (cycle_string.count(part1)*len(part1)==len(cycle_string)
            and len(part1)<len(cycle_string_min)):
            cycle_string_min=part1
    
    return cycle_string, cycle_string_min

getcontext().prec = 100
for d in range(1,50):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-5 (decimal 100, number 100개까지 확인)
# Issue 1: loop 7,13 등 오류 -> close -> new Issue 3.
# Issue 2: 순환아닌 부분반복 숫자 처리. loop 17,29 -> close
# Issue 3: 한자리 순환처리 -> close
# Issue 4 : loop 81 문제. 순환마디가 큰경우. -> close. decimal size increas

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string) #Issue -1
                and len(part1)*string.count(part1)>len(string)/2): #Issue-2
                cycle_string=part1
    #Issue-3 
    cycle_string_min=cycle_string
    for i in range(len(cycle_string)):
        part1=cycle_string[:i]
        if (cycle_string.count(part1)*len(part1)==len(cycle_string)
            and len(part1)<len(cycle_string_min)):
            cycle_string_min=part1
    
    return cycle_string, cycle_string_min

getcontext().prec = 100
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    print("{0} {1:100} {2}".format(d,string,len(string)),getReciprocal(string))
```

```python
# 메인루프-1-6 (decimal 1000, number 100까지)
# decimal 10,100,1000 별로 순환마디 체크/비교.

from decimal import *

def getReciprocal(string):
    #string=str(number)
    cycle_string=''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1=string[i:j]
            part2=string[j:j*2-i]
            if (part1 == part2 
                and len(part1)>len(cycle_string) #Issue -1
                and len(part1)*string.count(part1)>len(string)/2): #Issue-2
                cycle_string=part1
    #Issue-3 
    cycle_string_min=cycle_string
    for i in range(len(cycle_string)):
        part1=cycle_string[:i]
        if (cycle_string.count(part1)*len(part1)==len(cycle_string)
            and len(part1)<len(cycle_string_min)):
            cycle_string_min=part1
    return len(cycle_string_min)

getcontext().prec = 10
result=[]
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    result.append([d,getReciprocal(string)])
    if result[d-1][1] != 0:
        print(result[d-1])

getcontext().prec = 100
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    result[d-1].append(getReciprocal(string))
    if result[d-1][1] != result[d-1][2]:
        print(result[d-1])
        
getcontext().prec = 1000
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    result[d-1].append(getReciprocal(string))
    if result[d-1][2] != result[d-1][3]:
        print(result[d-1])
```

```python
# 메인루프-2 (decimal 100, number 100)
# 코드최적화,속도개선

from decimal import *

def getReciprocal(string):
    cycleMax = ''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1 = string[i:j]
            part2 = string[j:j*2-i]
            if part1 == part2:
                if len(part1) > len(cycleMax):
                    cycleMax = part1
    
    cycleMin = cycleMax
    for i in range(len(cycleMax)):
        part1 = cycleMax[:i]
        part2 = cycleMax[i:i*2]
        if part1 == part2:
            if len(part1)<len(cycleMin) and len(part1) != 0:
                cycleMin = part1
    return cycleMin

getcontext().prec = 100
result=[]
for d in range(1,100):
    string=str(Decimal(1)/Decimal(d))
    string=string[string.find(".")+1:]
    #result.append([d,getReciprocal(string)])
    #print(result[d-1])
    print("{0:2} {1:100} {2:2}".format(d,string,getReciprocal(string)))
```

```python
# 메인루프-2-2 (decimal 1000, number 1000)
# 일단 답부터 찾아보기
# 너무 느림,,

from decimal import *

def getReciprocal(string):
    cycleMax = ''
    for i in range(len(string)):
        for j in range(i,len(string)):
            part1 = string[i:j]
            part2 = string[j:j*2-i]
            if part1 == part2:
                if len(part1) > len(cycleMax):
                    cycleMax = part1
    
    cycleMin = cycleMax
    for i in range(len(cycleMax)):
        part1 = cycleMax[:i]
        part2 = cycleMax[i:i*2]
        if part1 == part2:
            if len(part1)<len(cycleMin) and len(part1) != 0:
                cycleMin = part1
    return len(cycleMin)

getcontext().prec = 1000
length = 0
maxLength = 0
for d in range(1,100):
    string = str(Decimal(1)/Decimal(d))
    string = string[string.find(".")+1:]
    length = getReciprocal(string)
    if length > maxLength:
        maxLength = length
    print(d,length,maxLength)
    #print("{0:2} {1:100} {2:2}".format(d,string,getReciprocal(string)))
print(maxLength)
```
