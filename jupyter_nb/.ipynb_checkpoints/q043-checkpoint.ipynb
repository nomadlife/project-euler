```python
# pandigit 수 만들기 -2 (속도 비교용, 32번에서 사용)
# 이방법이 20배 이상 빠름.

import time
start_time = time.time()

def get_next_pan(string):
    m=len(string)
    for i in range(m-1):
        if string[m-2-i]<string[m-1-i]:
            target_num = string[m-2-i]
            new_num = min([x for x in string[m-1-i:] if x>target_num])
            last_nums = ''.join(sorted(string[m-2-i:].replace(new_num,"")))
            result = string[:m-2-i] + new_num + last_nums
            return result

string = '0123456789'
while not string == '0132456789':
    string = str(string).zfill(10)
    string = get_next_pan(string)
    print(string, "pandigit")  
        
print("Calculation time",time.time() - start_time)
```

```python
# 최종계산

import time
start_time = time.time()

def get_next_pan(string):
    m=len(string)
    for i in range(m-1):
        if string[m-2-i]<string[m-1-i]:
            target_num = string[m-2-i]
            new_num = min([x for x in string[m-1-i:] if x>target_num])
            last_nums = ''.join(sorted(string[m-2-i:].replace(new_num,"")))
            result = string[:m-2-i] + new_num + last_nums
            return result

def is_divisible(string):
    prime=[2,3,5,7,11,13,17]
    for j in range(7,0,-1):
        if int(string[j:j+3])%prime[j-1] == 0:
            continue
        else:
            return False
    return True

total = 0
string = '0123456789'
while not string == None:
    if is_divisible(string):
        print(string, "divisible")
        total = total + int(string)
    string = get_next_pan(string)

print(total)
print("calculation time", time.time() - start_time)
```

```python
#forum-5
def p43(num=0, digits=[1]*10, passed=0):
    if passed == 10:
        num = str(num)
        for i in xrange(7):
            if int(num[i+1:i+4]) % (2,3,5,7,11,13,17)[i]:
                return 0
        return int(num)
    
    counter = 0
    for n in xrange(10):
        if digits[n]:
            digits[n] = 0
            counter += p43(num*10+n, digits, passed+1)
            digits[n] = 1

    return counter
```

```python
#forum-6
import itertools
import time


def property(num):
    numbers = [2, 3, 5, 7, 11, 13, 17]
    i = 1
    for number in numbers:
        if int(str(num)[i:i+3]) % number != 0:
            return False
        i += 1
    return True


def main():
    sum = 0
    for item in itertools.permutations('0123456789'):
        if item[0] != '0':
            if property(int(''.join(item))):
                sum += int(''.join(item))

    print(sum)


if __name__ == '__main__':
    start = time.time()
    main()
    print(time.time() - start)
```

```python
#forum-7
import time

def result():
    final_result    = []
    middle_4_digits = []
    last_5_digits   = []
    last_7_digits   = []
    last_8_digits   = []

    for i in range(1000, 100000):
        number_string_list_l = list(str(i))
        length_5_original    = len(number_string_list_l)
        length_5_arranged    = len(set(number_string_list_l))

        if length_5_original != length_5_arranged:
            continue

        elif i < 10000 and '0' in number_string_list_l:
            continue

        d_fgh = int(''.join(number_string_list_l[:-2]))
        d_ghi = int(''.join(number_string_list_l[-4:-1]))
        d_hij = int(''.join(number_string_list_l[-3:]))

        if d_fgh % 11 != 0 or d_ghi % 13 != 0 or d_hij % 17 != 0:
            continue

        if i < 10000:
            last_5_digits.append('0' + str(i))

        else:
            last_5_digits.append(str(i))

    for j in range(100, 10000):
        number_string_list_m = list(str(j))
        length_4_original    = len(number_string_list_m)
        length_4_arranged    = len(set(number_string_list_m))

        if length_4_original != length_4_arranged:
            continue

        elif j < 1000 and '0' in number_string_list_m:
            continue

        d_d   = int(''.join(number_string_list_m[0]))
        d_cde = int(''.join(number_string_list_m[:-2]))
        d_f   = int(''.join(number_string_list_m[-2]))
        d_efg = int(''.join(number_string_list_m[-3:]))

        if length_4_original == 4 and d_d % 2 != 0:
            continue

        elif d_cde % 3 != 0 or d_f % 5 != 0 or d_efg % 7 != 0:
            continue

        if j < 1000:
            middle_4_digits.append('0' + str(j))

        else:
            middle_4_digits.append(str(j))

    for x in range(len(middle_4_digits)):
        for y in range(len(last_5_digits)):
            if middle_4_digits[x][-2:] == last_5_digits[y][:2]:
                concatenated_string = middle_4_digits[x] + last_5_digits[y][2:]

                if len(concatenated_string) == len(set(concatenated_string)):
                    last_7_digits.append(concatenated_string)

    for m in range(10):
        for n in range(len(last_7_digits)):
            temporary1 = str(m) + last_7_digits[n]

            length_temporary1_original = len(list(temporary1))
            length_temporary1_arranged = len(list(set(temporary1)))

            if length_temporary1_original != length_temporary1_arranged:
                continue

            if int(temporary1[:3]) % 3 != 0:
                continue

            last_8_digits.append(str(m) + last_7_digits[n])

    for p in range(10, 100):
        for q in range(len(last_8_digits)):
            temporary2 = str(p) + last_8_digits[q]

            length_temporary2_original = len(list(temporary2))
            length_temporary2_arranged = len(list(set(temporary2)))

            if length_temporary2_original != length_temporary2_arranged:
                continue

            final_result.append(int(str(p) + last_8_digits[q]))

    return sum(final_result)

start_time = time.time()

print(result())
print("{} ms".format(round(1000 * (time.time() - start_time))))
```

```python
#forum-9
import time

def digits(x):
    return list(map(int, str(x)))

def validMultiple(x):
     return (len(set(digits(x))) == 3 and x > 100) or (x < 100)

solutions = []
factors = [13, 11, 7, 5, 3, 2]


def findSolutions(substring: str):
 
    unused_digits = [x for x in range(0,10) if str(x) not in substring]
    
    # d1 is irrelevant
    if len(unused_digits) == 1:
        chunk = str(unused_digits[0]) + substring
        solutions.append(chunk)
        return

    # 17 is not in factors therefore we have to subtract 3
    f = factors[len(substring) - 3]
    for x in unused_digits:
        chunk = str(x) + substring     
        if int(chunk[:3]) % f == 0:
            findSolutions(chunk)



def solve():
    global solutions
    solutions = []
    multiples = [x for x in range(1,1000) if x % 17 == 0]
    valid_multiples = filter(validMultiple, multiples)

    for x in valid_multiples:
        findSolutions(str(x))
    return sum(map(int, solutions))


if __name__ == "__main__":
    start = time.time()
    result = solve()
    end = time.time()
    print("Result: ", result, "\nTime elapsed: ", (end-start))
```

```python
#forum-10
from itertools import permutations

# Numbers are represented as strings to ease digit selection
def has_property(number):
    c1 = int(number[1:4]) % 2 == 0
    c2 = int(number[2:5]) % 3 == 0
    c3 = int(number[3:6]) % 5 == 0
    c4 = int(number[4:7]) % 7 == 0
    c5 = int(number[5:8]) % 11 == 0
    c6 = int(number[6:9]) % 13 == 0
    c7 = int(number[7:10]) % 17 == 0
    return c1 and c2 and c3 and c4 and c5 and c6 and c7

numbers = [''.join(number) for number in list(permutations('0123456789'))]
print(sum(int(number) for number in numbers if has_property(number)))
```

```python
#forum-11
import time
from itertools import permutations
primes = [2,3,5,7,11,13,17]
def main():
    sum = 0
    for k in permutations("0123456789"):
        ks = ""
        for c in k:
            ks+=c
        #print(ks)
        found = True
        for i in range(1, 8):
            if int(ks[i:i+3])%primes[i-1] != 0:
                found = False
                continue
        if found:
            print("-" + ks)
            sum+=int(ks)
    print(sum)
start_time = time.time()
main()
print("Elapsed time: {}ms".format(round(1000*(time.time() - start_time))))
```

```python
#forum-13
def to_3c(n):
    s = str(n)
    return "0" * (3 - len(s)) + s

def recur(n = 0):
    if n == 0:
        return sum([int(s) for s in recur(n + 1)])
    elif n == 7:
        return [to_3c(n) for n in range(17, 1000, 17)]
    else:
        result = []
        for i in recur(n + 1):
            for x in ({x for x in "0123456789"} - set(i)):
                new = x + i
                if (int(new[:3]) % (2, 3, 5, 7, 11, 13)[n - 1] == 0
                    and len(new) == len(set(new))):
                    result.append(new)

        return result

print(recur())
```
