## Poker hands




```python
with open('p054_poker.txt') as data:
    lines = data.readlines()

cards=[]
for line in lines:
    Arr = line.split()
    p1n,p1s,p2n,p2s ='','','',''
    for i in range(5):
        p1n += Arr[i][0]
        p1s += Arr[i][1]
        p2n += Arr[i+5][0]
        p2s += Arr[i+5][1]
    cards.append([p1n,p1s,p2n,p2s])
print(cards)
```

```python

def checkCards(numbers, suits):
    newStr =''
    result = ''
    for i in '23456789TJQKA':
        count = numbers.count(i)
        if count > 0: 
            if count == 2: result += 'P'+i
            elif count == 3: result += 'T'+i
            elif count == 4 : result += 'F'+i
            newStr += i*count
    if newStr in '23456789TJQKA':
        result += 'Straight'
    if all(suits[i] == suits[i+1] for i in range(4)): 
        result += 'Flush'
    return result
print(determine('77733','HDSDH'))
```
